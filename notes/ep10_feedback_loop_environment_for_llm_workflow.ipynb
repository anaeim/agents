

An agent can have a feedback loop.<br>
the feekback envriroment rely on feedback loops, where information can cycle back and be processed multiple times.

The key difference is that theres no fixed sequence of steps. Instead of a predefined path, the process is fluid and dynamic. This flexibility makes agents much more powerful, capable of tackling a broader range of problems. But it also introduces trade-offs: theyre less predictable, and raise important concerns about robustness, guardrails, and reliability. If you want an agent system to run in production, you need to think carefully about constraints like guaranteed completion and execution time.

The diagram here illustrates this more generic agent loop. Unlike earlier patterns where we had clear inputs and outputs, here the output is really an interaction with an environment. A human can make a request, which goes to the LLM. The LLM then takes an action in the environment, receives feedback, and decides what to do next.

This loop continuesaction, feedback, decisionuntil the LLM decides its done and stops.
