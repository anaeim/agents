# OpenAI Agents SDK  Core Terminology

##  Agents
An **agent** is a package around to call LLMs for a purpose.  
It is the core unit you create and run when working with the OpenAI Agents SDK.

---

##  Handoff
A **handoff** is the name used for **interactions between agents**.  
This concept appears in different contexts but always refers to how agents communicate or pass work to one another.

---

##  Guardrails
**Guardrails** are the **checks and controls** you set up to ensure an agent is doing what you want it to do.  
They stop the agent from "going off the rails" and keep its behavior within safe, expected boundaries.


#  Steps to Run an Agent

1. **Create an instance of an agent**  
   - This defines the role or purpose of the agent in your solution.

2. **Use `with_trace`**  
   - (Optional but recommended)  
   - This keeps a log of all interactions with the agent, which can be viewed in OpenAIs monitoring tools.

3. **Call `runner.run`**  
   - This is the async function (coroutine) that actually runs the agent.  
   - Youll need to `await runner.run` to execute it properly.
