## 2. tools





Tools is when you give an llm power and authority to execute actions on your behalf.

##  Tool Calling

At its core, **tool calling** is about getting an LLM to produce **structured JSON output**.  
The model doesnt perform the action directlyinstead, it specifies what it wants done, along with any parameters, in JSON.  

###  How it works
1. **LLM outputs JSON**  
   - The model produces a structured request describing the tool/action and its inputs.  

2. **Code interprets the request**  
   - Your orchestration layer (often just an `if` / `match` statement) decides which tool to run.  

3. **Execute the action**  
   - The corresponding function or API is called with the provided parameters.  

4. **Return results to the LLM**  
   - The outcome is passed back into the model, allowing it to continue reasoning with the new information.  

###  Key idea
From the LLMs perspective, it looks like its calling a tool.  
In reality, its **your code** that executes the action and feeds the result back.
